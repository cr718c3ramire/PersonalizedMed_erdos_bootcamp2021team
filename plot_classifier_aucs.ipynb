# Introduction
This notebook visualizes the results from our classification task. We will see how the encoder and classifier affects performance on classifying scientific text into various categories.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
aucs = pd.read_excel('Classifier_AUCs.xlsx')
aucs.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(x="class",y='AUC',hue='encoder',data=aucs)
ax.set_xlabel('Class', fontsize=16)
ax.set_ylabel('Area Under ROC', fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(x="class",y='AUC',hue='classifier',data=aucs)
ax.set_xlabel('Class', fontsize=16)
ax.set_ylabel('Area Under ROC', fontsize=16)
plt.show()
```

## Summary
* Doc2Vec encoder outperforms the BoW and TF-IDF. 
* Random forest model slightly outperforms the other models. 
* Interestingly, the classifier used had less effect on the classification performance than the encoder.
